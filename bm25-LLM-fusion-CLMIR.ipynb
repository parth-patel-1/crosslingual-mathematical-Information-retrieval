```python
# create dataframe from the result file
def load_dataframes(query_file, passage_file, result_file, top_k=100):
    query_df = pd.read_csv(query_file, sep='\t', header=None, names=['query_id', 'query_text'])
    passage_df = pd.read_csv(passage_file, sep='\t', header=None, names=['doc_id', 'doc_text'])

    result_df = pd.read_csv(result_file, sep=' ', header=None,
                            names=['query_id', '_', 'doc_id', 'rank', 'score', '__'])

    # Keep top-k ranked results
    result_df = result_df[result_df['rank'] <= top_k]

    result_df = result_df[['query_id','doc_id','score']]
    
    # Merge query & passage with results
    merged_df = result_df.merge(query_df, on='query_id', how='left')
    merged_df = merged_df.merge(passage_df, on='doc_id', how='left')

    return merged_df
```

```python
from tqdm import tqdm
import pandas as pd
from huggingface_hub import login
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import torch.nn.functional as F

# Login (only required once per session)
login(token="HF_TOKEN")

# ✅ Load Gemma Model (2B or 7B)
model_name = "google/gemma-3-12b-it"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,  # use float16 if your GPU doesn't support bfloat16
    device_map="auto"
)
model.eval()
```

```python
import pandas as pd

def pointwise_score(query, doc):
    prompt = f"""नीचे एक हिंदी गणित प्रश्न (Query) और एक हिंदी गणितीय दस्तावेज़ (Document) दिया गया है। यहाँ दिया गया दस्तावेज़ प्रश्न से सम्बंधित है या नहीं — उसके आधार पर केवल "Yes" या "No" लिखिए, कोई अन्य शब्द न लिखें।

प्रारूप:
Query: {query}
Document: {doc}

Answer:"""

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=1,
        return_dict_in_generate=True,
        output_scores=True,
        output_logits=True,
        do_sample=False,
        pad_token_id=tokenizer.eos_token_id
    )

    # Get token IDs
    id_yes = tokenizer(" Yes", add_special_tokens=False).input_ids[0]
    id_no = tokenizer(" No", add_special_tokens=False).input_ids[0]

    # Extract logits for YES/NO from the first generation step
    logits = outputs.scores[0][0]              # tensor of shape [vocab_size]
    selected = torch.stack([logits[id_yes], logits[id_no]])  # shape [2]

    # Apply softmax to get probabilities
    probs = F.softmax(selected, dim=0)         # a 2-element tensor
    yes_prob = probs[0].item()
    prediction = "Yes" if yes_prob > probs[1].item() else "No"

    return yes_prob, prediction

def process_all_df(query_document_df):
    scores = []
    outputs = []
    for _, row in tqdm(query_document_df.iterrows(), total=len(query_document_df)):
        score, output = pointwise_score(row['query_text'], row['doc_text'])
        scores.append(score)
        outputs.append(output)

    query_document_df['relevance_score'] = scores
    query_document_df['output'] = outputs

    return query_document_df[['query_id', 'doc_id', 'query_text', 'doc_text','score','relevance_score', 'output']]
```
